### Process .PAL lut file to c array

```python
import ipywidgets as widgets
```

```python
fileNameInput = widgets.Text(
    value='*.pal',
    placeholder='Type pal file',
    description='file:',
    disabled=False
)
processButton = widgets.Button(
    value=False,
    description='Process',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Description',
    icon='check'
)
placeHolder = widgets.VBox([fileNameInput,processButton])
#b,g,r,a   low->high
def ProcessPAL(b):
    fileName = fileNameInput.value
    with open(fileName,'r') as file:

        allData = file.readlines()
        colorNums = int(allData[2])
        colorList = []
        print(colorNums)
        for i in range(colorNums):
            currentLine = allData[i+3]
            currentLine = currentLine.split(' ')
            red=int(currentLine[0])
            green=int(currentLine[1])
            blue=int(currentLine[2])
            colorList.append((red<<16)|(green<<8)|(blue)|(0xff<<24))
    for i in range(colorNums):
        print('%#08X'%(colorList[i]))
    with open(fileName[:-3]+'c','w') as outputFile: 
        outputFile.write('//Order is b g r a, low->high\n')
        outputFile.write('const uint32_t pal['+str(colorNums)+']={\n')
        for i in range(colorNums):
            outputFile.write('\t%#08X,\n'%(colorList[i]))
        outputFile.write('};\n')
        
processButton.on_click(ProcessPAL)
placeHolder
```

```python
from PIL import Image

pngNameInput = widgets.Text(
    value='*.png',
    placeholder='Type png file',
    description='file:',
    disabled=False
)
processButton = widgets.Button(
    value=False,
    description='Process',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Description',
    icon='check'
)
def ProcessImg(Img):
    pngName = pngNameInput.value

    im = Image.open(pngName)
    x,y = im.size
    testPal = im.getpalette()    
    imData = im.load()
    indexVal =[]
    for i in range(y):
        for j in range(x):
            lowPixel = imData[j,i]
            indexVal.append(lowPixel)
    #print(indexVal)
    dataLength = len(indexVal)
    with open(pngName[:-3]+'c','w') as outputFile: 
        outputFile.write('//image size %d ,%d\n'%(x,y))
        outputFile.write('const uint8_t img['+str(dataLength)+']={\n')
        left = dataLength%16
        for i in range(int(dataLength/16)):
            outputFile.write('\t')
            for j in range(16):
                outputFile.write('%#02X, '%indexVal[16*i+j])
            outputFile.write('\n')
        if left != 0:
            for k in range(left):
                outputFile.write('\t')
                outputFile.write('%#02X, '%indexVal[int(dataLength/16)+k])
            outputFile.write('\n')    
        outputFile.write('};\n')        
processButton.on_click(ProcessImg)
placeHolder = widgets.VBox([pngNameInput,processButton])
placeHolder
```
